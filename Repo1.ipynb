# **Eksperimen Data Mining Dengan One-Hot Encoding**
Prediksi Minat Beli Produk Berdasarkan Profil Pelanggan

One-Hot Encoding?

Teknik mengubah data kategori (teks) menjadi angka biner (0 dan 1)

Contoh: "Jakarta", "Bandung", "Surabaya" → [1,0,0], [0,1,0], [0,0,1]

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#data contoh
data = {
    'usia': [25, 35, 45, 23, 40, 28, 50, 33, 29, 42],
    'kota': ['Jakarta', 'Bandung', 'Surabaya', 'Jakarta', 'Bandung',
             'Surabaya', 'Jakarta', 'Bandung', 'Surabaya', 'Jakarta'],
    'pekerjaan': ['Mahasiswa', 'Karyawan', 'Wiraswasta', 'Mahasiswa', 'Karyawan',
                  'Wiraswasta', 'Karyawan', 'Wiraswasta', 'Mahasiswa', 'Karyawan'],
    'minat_beli': [0, 1, 1, 0, 1, 0, 1, 1, 0, 1]  # 0=Tidak, 1=Ya
}

df = pd.DataFrame(data)
print("=== DATA ASLI ===")
print(df)
print("\n")
```

Dataset:

- Usia: Data numerik (sudah bisa dipahami komputer)

- Kota & Pekerjaan: Data kategori (perlu diubah ke angka)

- Minat Beli: Target yang ingin kita prediksi









```python
# Ubah data kategori menjadi format one-hot
kolom_kategorikal = ['kota', 'pekerjaan']
df_encoded = pd.get_dummies(df, columns=kolom_kategorikal, prefix=kolom_kategorikal)
df_encoded = df_encoded.astype(int)

print("Data Setelah One-Hot Encoding:")
print(df_encoded)
```

Model Prediksi

```python
# Pisahkan data: Features (X) dan Target (y)
X = df_encoded.drop('minat_beli', axis=1)  # Semua kolom kecuali minat_beli
y = df_encoded['minat_beli']               # Hanya kolom minat_beli

print("Features (X) untuk training:")
print(X.columns.tolist())
```

Konsep Features & Target:

- Features (X): Data yang digunakan untuk memprediksi (usia, kota, pekerjaan)

- Target (y): Hasil yang ingin diprediksi (minat_beli)



Training Model Machine Learning

```python
# Bagi data menjadi training (70%) dan testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat model Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)  # Proses belajar dari data training

print("✅ Model selesai belajar dari data!")
```

```python
# Test model dengan data testing
y_pred = model.predict(X_test)
akurasi = accuracy_score(y_test, y_pred)

print("=== HASIL EVALUASI ===")
print(f"Akurasi model: {akurasi:.2f} ({(akurasi*100):.0f}%)")

print("\nPerbandingan Hasil:")
test_results = pd.DataFrame({
    'Aktual': y_test.values,
    'Prediksi': y_pred
})
print(test_results)
```

Interpretasi Hasil:


- Model bisa memprediksi dengan sempurna pada data testing

- Dalam real world, biasanya akurasi tidak 100% sempurna

```python
# Data pelanggan baru yang ingin diprediksi
data_baru = {
    'usia': [30, 22, 48],
    'kota': ['Jakarta', 'Bandung', 'Surabaya'],
    'pekerjaan': ['Karyawan', 'Mahasiswa', 'Wiraswasta']
}

df_baru = pd.DataFrame(data_baru)

# Lakukan one-hot encoding yang sama
df_baru_encoded = pd.get_dummies(df_baru, columns=kolom_kategorikal)

# Prediksi minat beli
prediksi_baru = model.predict(df_baru_encoded)

# Tampilkan hasil
df_baru['prediksi_minat_beli'] = prediksi_baru
df_baru['keterangan'] = df_baru['prediksi_minat_beli'].map({0: 'Tidak Minat', 1: 'Minat'})

print("=== PREDIKSI PELANGGAN BARU ===")
print(df_baru)
```

Hasil Prediksi:

Usia 30, Jakarta, Karyawan → Minat

Usia 22, Bandung, Mahasiswa → Tidak Minat

Usia 48, Surabaya, Wiraswasta → Minat

```python
# Insight yang Didapat
# Lihat faktor paling berpengaruh
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

print("Faktor paling berpengaruh dalam minat beli:")
print(feature_importance)
```

Kesimpulan:

1. One-Hot Encoding berhasil mengubah data kategori menjadi format numerik

2. Model machine learning bisa belajar pola dari data yang sudah di-encode

3. Hasil prediksi bisa digunakan untuk mengetahui minat beli pelanggan baru

Ringkasan Proses:

Data Kategori → One-Hot Encoding → Data Numerik → Machine Learning → Prediksi
